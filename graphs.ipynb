```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from CSV
flags = ["O0"]
sizes = ["small","med","big"]
elem_num = { "small": 100 , "med" : 2500, "big": 5000}
# Extract relevant columns
for flag in flags:
    for size in sizes:
        data_path = f"csvData/{flag}/{size}{flag}Sorted.csv"
        data = pd.read_csv(data_path)
        programs = data['Program']
        powerlimit = data['PowerLimit']
        powerconsumed = data['Package']
        time = data['Time']
        memory = data['Memory']
        # Create DataFrame
        df = pd.DataFrame({'Program': programs, 'PowerLimit': powerlimit, 'Time': time})
        # Calculate mean time and memory for each power limit
        mean_data = df.groupby(['Program','PowerLimit']).mean()
        # Plot
        mean_data.plot(kind='bar', figsize=(10, 6))
        plt.figure(figsize=(10, 6))
        sns.barplot(data=df, x='PowerLimit', y='Time', hue='Program')
        plt.title(f"Runtime of Each Algorithm ({flag},{elem_num[size]} elems) Sorted")
        plt.xlabel('Power Limit')
        plt.ylabel('Runtime (ms)')
        plt.grid(axis='y')
        plt.legend(title='Algorithm')
        plt.savefig(f"images/runtime{flag}{size}Sorted.png")
        #plt.show()
    
```
